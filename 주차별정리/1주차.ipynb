````python
# 1-2. enumerateì™€ enumerate í™œìš©
arr = ['a', 'b', 'c', 'd', 'e']

print("=== enumerate í™œìš© ===")
for i, val in enumerate(arr):
    print(f"ì¸ë±ìŠ¤ {i}: {val}")

print("\n=== ì¡°ê±´ì— ë§ëŠ” ì¸ë±ìŠ¤ ì°¾ê¸° ===")
target = 'c'
for i, val in enumerate(arr):
    if val == target:
        print(f"'{target}'ëŠ” ì¸ë±ìŠ¤ {i}ì— ìˆìŠµë‹ˆë‹¤.")
        break

# enumerateëŠ” ì‹œì‘ ê°’ì„ ì§€ì •í•  ìˆ˜ ìˆìŒ
print("\n=== enumerate with start ===")
for i, val in enumerate(arr, start=1):
    print(f"{i}ë²ˆì§¸: {val}")

print(list(map(lambda x: (x[0], x[1]), enumerate(arr))))

print("\n=== Lambda í•¨ìˆ˜ ì™„ì „ ì •ë³µ ===")
print("ğŸ’¡ Lambda ê¸°ë³¸ ë¬¸ë²•: lambda ë§¤ê°œë³€ìˆ˜: ë°˜í™˜ê°’")

# 1. ê¸°ë³¸ lambda ì˜ˆì œ
print("\n1ï¸âƒ£ ê¸°ë³¸ lambda í•¨ìˆ˜:")
# ì¼ë°˜ í•¨ìˆ˜ vs lambda í•¨ìˆ˜
def square_normal(x):
    return x ** 2

square_lambda = lambda x: x ** 2

print(f"ì¼ë°˜ í•¨ìˆ˜: square_normal(5) = {square_normal(5)}")
print(f"ëŒë‹¤ í•¨ìˆ˜: square_lambda(5) = {square_lambda(5)}")

# 2. map()ê³¼ lambda ì¡°í•©
print("\n2ï¸âƒ£ map()ê³¼ lambda ì¡°í•©:")
numbers = [1, 2, 3, 4, 5]

# ê° ìˆ«ìë¥¼ ì œê³±
squares = list(map(lambda x: x ** 2, numbers))
print(f"ì œê³±: {numbers} â†’ {squares}")

# ê° ìˆ«ìì— 10 ë”í•˜ê¸°
plus_ten = list(map(lambda x: x + 10, numbers))
print(f"+10: {numbers} â†’ {plus_ten}")

# 3. enumerate + lambda ì‹¤ìš© ì˜ˆì œ
print("\n3ï¸âƒ£ enumerate + lambda ì‹¤ìš© ì˜ˆì œ:")
fruits = ['apple', 'banana', 'cherry']

# ì›ë³¸ ì½”ë“œ (ì˜ë¯¸ ì—†ëŠ” ë³€í™˜)
original = list(map(lambda x: (x[0], x[1]), enumerate(fruits)))
print(f"ì›ë³¸ ì½”ë“œ ê²°ê³¼: {original}")

# ì‹¤ìš©ì ì¸ ì˜ˆì œë“¤
# ì˜ˆì œ 1: ì¸ë±ìŠ¤ì™€ ê°’ì„ ë¬¸ìì—´ë¡œ ì¡°í•©
indexed_fruits = list(map(lambda x: f"{x[0]}: {x[1]}", enumerate(fruits)))
print(f"ì¸ë±ìŠ¤ ì¡°í•©: {indexed_fruits}")

# ì˜ˆì œ 2: ì¡°ê±´ë¶€ ë³€í™˜ (ì§ìˆ˜ ì¸ë±ìŠ¤ë§Œ ëŒ€ë¬¸ì)
conditional = list(map(lambda x: x[1].upper() if x[0] % 2 == 0 else x[1], enumerate(fruits)))
print(f"ì§ìˆ˜ ì¸ë±ìŠ¤ ëŒ€ë¬¸ì: {conditional}")

# ì˜ˆì œ 3: ë”•ì…”ë„ˆë¦¬ ìƒì„±
fruit_dict = dict(map(lambda x: (x[1], x[0]), enumerate(fruits)))
print(f"ë”•ì…”ë„ˆë¦¬ ìƒì„±: {fruit_dict}")

print("\n4ï¸âƒ£ filter()ì™€ lambda:")
# ì§ìˆ˜ë§Œ í•„í„°ë§
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(f"ì§ìˆ˜ í•„í„°ë§: {numbers} â†’ {evens}")

# ê¸¸ì´ê°€ 5 ì´ìƒì¸ ê³¼ì¼ë§Œ
long_fruits = list(filter(lambda x: len(x) >= 5, fruits))
print(f"ê¸´ ê³¼ì¼ëª…: {fruits} â†’ {long_fruits}")

print("\n5ï¸âƒ£ sorted()ì™€ lambda:")
students = [('Alice', 85), ('Bob', 90), ('Charlie', 78)]

# ì ìˆ˜ë¡œ ì •ë ¬
by_score = sorted(students, key=lambda x: x[1])
print(f"ì ìˆ˜ìˆœ ì •ë ¬: {by_score}")

# ì´ë¦„ ê¸¸ì´ë¡œ ì •ë ¬
by_name_len = sorted(students, key=lambda x: len(x[0]))
print(f"ì´ë¦„ ê¸¸ì´ìˆœ: {by_name_len}")

print("\n6ï¸âƒ£ ë³µì¡í•œ lambda ì˜ˆì œ:")
# ì—¬ëŸ¬ ë§¤ê°œë³€ìˆ˜
add = lambda x, y: x + y
print(f"ë‘ ìˆ˜ ë”í•˜ê¸°: add(3, 5) = {add(3, 5)}")

# ì¡°ê±´ë¶€ í‘œí˜„ì‹ (ì‚¼í•­ ì—°ì‚°ì)
abs_lambda = lambda x: x if x >= 0 else -x
print(f"ì ˆëŒ“ê°’: abs_lambda(-5) = {abs_lambda(-5)}")

# ì¤‘ì²© lambda (ê¶Œì¥í•˜ì§€ ì•ŠìŒ)
multiply_add = lambda x: (lambda y: x * y + 1)
result = multiply_add(3)(4)  # 3 * 4 + 1 = 13
print(f"ì¤‘ì²© lambda: {result}")

print("\n7ï¸âƒ£ ì½”ë”©í…ŒìŠ¤íŠ¸ì—ì„œ lambda í™œìš©:")
# ì¢Œí‘œ ì •ë ¬ (ê±°ë¦¬ìˆœ)
points = [(1, 2), (3, 1), (0, 0), (2, 3)]
by_distance = sorted(points, key=lambda p: p[0]**2 + p[1]**2)
print(f"ì›ì ì—ì„œ ê±°ë¦¬ìˆœ: {by_distance}")

# ë¬¸ìì—´ ì •ë ¬ (ê¸¸ì´ â†’ ì‚¬ì „ìˆœ)
words = ['apple', 'pie', 'banana', 'cat']
sorted_words = sorted(words, key=lambda w: (len(w), w))
print(f"ê¸¸ì´â†’ì‚¬ì „ìˆœ: {sorted_words}")

print("\nğŸ’¡ Lambda ì‚¬ìš© íŒ:")
print("âœ… ê°„ë‹¨í•œ ì¼íšŒìš© í•¨ìˆ˜ì— ì‚¬ìš©")
print("âœ… map, filter, sortedì˜ key í•¨ìˆ˜ë¡œ í™œìš©")
print("âŒ ë³µì¡í•œ ë¡œì§ì€ ì¼ë°˜ í•¨ìˆ˜ ì‚¬ìš©")
print("âŒ ê°€ë…ì„±ì´ ë–¨ì–´ì§€ë©´ ì¼ë°˜ í•¨ìˆ˜ ê³ ë ¤")
````

````python
# 1-3. joinê³¼ split í™œìš©
text = "apple,banana,cherry,date"
words = ["Hello", "World", "Python", "Programming"]

print("=== split í™œìš© ===")
fruits = text.split(',')
print(f"split ê²°ê³¼: {fruits}")

print("\n=== join í™œìš© ===")
# ë¦¬ìŠ¤íŠ¸ë¥¼ ë¬¸ìì—´ë¡œ í•©ì¹˜ê¸°
sentence = ' '.join(words)
print(f"join ê²°ê³¼: {sentence}")

# ë¬¸ìì—´ ëˆ„ì í•  ë•Œ íš¨ìœ¨ì ì¸ ë°©ë²•
print("\n=== ë¬¸ìì—´ ëˆ„ì  ë¹„êµ ===")
# ë¹„íš¨ìœ¨ì ì¸ ë°©ë²• (ë¬¸ìì—´ì€ ë¶ˆë³€ì´ë¯€ë¡œ ë§¤ë²ˆ ìƒˆë¡œìš´ ë¬¸ìì—´ ìƒì„±)
result1 = ""
for word in words:
    result1 += word + " "
print(f"ë¹„íš¨ìœ¨ì  ë°©ë²•: '{result1.strip()}'")

# íš¨ìœ¨ì ì¸ ë°©ë²• (ë¦¬ìŠ¤íŠ¸ì— append í›„ join)
parts = []
for word in words:
    parts.append(word)
result2 = ' '.join(parts)
print(f"íš¨ìœ¨ì  ë°©ë²•: '{result2}'")

print("\n=== ì‹¤ì œ ì½”ë”©í…ŒìŠ¤íŠ¸ join í™œìš© ì˜ˆì‹œ ===")

# ì˜ˆì‹œ 1: ìˆ«ì ë°°ì—´ì„ ë¬¸ìì—´ë¡œ ë³€í™˜
numbers = [1, 2, 3, 4, 5]
number_string = ''.join(map(str, numbers))
print(f"ìˆ«ì ë°°ì—´ì„ ë¬¸ìì—´ë¡œ: {numbers} -> '{number_string}'")

# ì˜ˆì‹œ 2: ì´ì§„ìˆ˜ ë§Œë“¤ê¸°
binary_bits = [1, 0, 1, 1, 0, 1]
binary_string = ''.join(map(str, binary_bits))
print(f"ì´ì§„ìˆ˜ ë§Œë“¤ê¸°: {binary_bits} -> '{binary_string}'")

# ì˜ˆì‹œ 3: ê²½ë¡œ ë§Œë“¤ê¸° (ìŠ¬ë˜ì‹œë¡œ ì—°ê²°)
path_parts = ["home", "user", "documents", "file.txt"]
file_path = '/'.join(path_parts)
print(f"ê²½ë¡œ ë§Œë“¤ê¸°: {path_parts} -> '{file_path}'")

# ì˜ˆì‹œ 4: SQL ì¿¼ë¦¬ ì¡°ê±´ ë§Œë“¤ê¸°
conditions = ["age > 18", "city = 'Seoul'", "status = 'active'"]
where_clause = ' AND '.join(conditions)
print(f"SQL ì¡°ê±´: {conditions} -> '{where_clause}'")

# ì˜ˆì‹œ 5: ë‹¨ì–´ ì¡°í•©í•˜ê¸° (ì•„ë‚˜ê·¸ë¨, ìˆœì—´ ë¬¸ì œ)
chars = ['a', 'b', 'c']
word = ''.join(chars)
print(f"ë¬¸ì ì¡°í•©: {chars} -> '{word}'")

# ì˜ˆì‹œ 6: ë¬¸ìì—´ ë’¤ì§‘ê¸° í›„ í•©ì¹˜ê¸°
reversed_chars = ['o', 'l', 'l', 'e', 'H']
reversed_word = ''.join(reversed_chars)
print(f"ë’¤ì§‘íŒ ë¬¸ì í•©ì¹˜ê¸°: {reversed_chars} -> '{reversed_word}'")

# ì˜ˆì‹œ 7: íŠ¹ì • ì¡°ê±´ì˜ ë¬¸ìë§Œ í•©ì¹˜ê¸° (í•„í„°ë§ + join)
mixed_data = ['a', '1', 'b', '2', 'c', '3']
letters_only = ''.join(c for c in mixed_data if c.isalpha())
numbers_only = ''.join(c for c in mixed_data if c.isdigit())
print(f"ë¬¸ìë§Œ: '{letters_only}', ìˆ«ìë§Œ: '{numbers_only}'")

# ì˜ˆì‹œ 8: í–‰ë ¬ì„ ë¬¸ìì—´ë¡œ ì¶œë ¥ (2D -> 1D)
matrix = [['1', '2', '3'], ['4', '5', '6'], ['7', '8', '9']]
matrix_string = '\n'.join(''.join(row) for row in matrix)
print(f"í–‰ë ¬ì„ ë¬¸ìì—´ë¡œ:\n{matrix_string}")

print("\nğŸ’¡ join ì‚¬ìš© íŒ:")
print("- ë¬¸ìì—´ ëˆ„ì : ''.join(list) ì‚¬ìš©")
print("- ê²½ë¡œ/URL: '/'.join() ë˜ëŠ” '.'.join() ì‚¬ìš©") 
print("- SQL/ì¡°ê±´: ' AND '.join() ì‚¬ìš©")
print("- ì¶œë ¥ í˜•ì‹: '\\n'.join() ë˜ëŠ” ', '.join() ì‚¬ìš©")
````

````python
# 1-4. ì•ŒíŒŒë²³ ì¹´ìš´íŒ…ê³¼ ë¬¸ìì—´ ì²˜ë¦¬
s = "Hello World Python Programming"

print("=== ord()ì™€ chr() í•¨ìˆ˜ ì´í•´í•˜ê¸° ===")
# ord(): ë¬¸ì â†’ ASCII ì½”ë“œ (ìˆ«ì)
# chr(): ASCII ì½”ë“œ (ìˆ«ì) â†’ ë¬¸ì
print(f"ord('a') = {ord('a')}")  # 97
print(f"ord('b') = {ord('b')}")  # 98  
print(f"ord('z') = {ord('z')}")  # 122
print(f"ord('A') = {ord('A')}")  # 65
print(f"chr(97) = '{chr(97)}'")  # 'a'
print(f"chr(98) = '{chr(98)}'")  # 'b'

print(f"\n=== ì•ŒíŒŒë²³ ìœ„ì¹˜ ê³„ì‚° ì›ë¦¬ ===")
# 'a'ëŠ” 0ë²ˆì§¸, 'b'ëŠ” 1ë²ˆì§¸, ..., 'z'ëŠ” 25ë²ˆì§¸ ì•ŒíŒŒë²³
example_char = 'h'
position = ord(example_char) - ord('a')
print(f"'{example_char}'ì˜ ì•ŒíŒŒë²³ ìœ„ì¹˜: ord('{example_char}') - ord('a') = {ord(example_char)} - {ord('a')} = {position}")
print(f"ì¦‰, '{example_char}'ëŠ” {position}ë²ˆì§¸ ì•ŒíŒŒë²³ì…ë‹ˆë‹¤")

print(f"\n=== ì•ŒíŒŒë²³ ì¹´ìš´íŒ… (ëŒ€ì†Œë¬¸ì êµ¬ë¶„ ì—†ì´) ===")
# ë°©ë²• 1: ë°°ì—´ì„ ì´ìš©í•œ ì¹´ìš´íŒ… (ê°€ì¥ íš¨ìœ¨ì )
cnt = [0] * 26  # a~zê¹Œì§€ 26ê°œ ì•ŒíŒŒë²³ì„ ìœ„í•œ ë°°ì—´

print("ì¹´ìš´íŒ… ê³¼ì •:")
for c in s.lower():
    if 'a' <= c <= 'z':  # ì•ŒíŒŒë²³ì¸ì§€ í™•ì¸
        index = ord(c) - ord('a')  # ì•ŒíŒŒë²³ ìœ„ì¹˜ ê³„ì‚°
        cnt[index] += 1
        print(f"  '{c}' â†’ ì¸ë±ìŠ¤ {index} â†’ cnt[{index}] = {cnt[index]}")

print(f"\nìµœì¢… ì¹´ìš´íŒ… ë°°ì—´: {cnt}")
print("ê²°ê³¼ (0ì´ ì•„ë‹Œ ê²ƒë§Œ):")
for i in range(26):
    if cnt[i] > 0:
        alphabet = chr(ord('a') + i)  # ì¸ë±ìŠ¤ë¥¼ ë‹¤ì‹œ ë¬¸ìë¡œ ë³€í™˜
        print(f"  '{alphabet}': {cnt[i]}ê°œ")

print("\n=== ë‹¤ë¥¸ ì¹´ìš´íŒ… ë°©ë²•ë“¤ê³¼ ë¹„êµ ===")
# ë°©ë²• 2: ë”•ì…”ë„ˆë¦¬ ì‚¬ìš© (ë” ì§ê´€ì ì´ì§€ë§Œ ì•½ê°„ ëŠë¦¼)
from collections import Counter

dict_count = {}
for c in s.lower():
    if 'a' <= c <= 'z':
        dict_count[c] = dict_count.get(c, 0) + 1

print("ë”•ì…”ë„ˆë¦¬ ë°©ë²•:", dict_count)

# ë°©ë²• 3: Counter ì‚¬ìš© (ê°€ì¥ ê°„ë‹¨)
counter_result = Counter(c for c in s.lower() if 'a' <= c <= 'z')
print("Counter ë°©ë²•:", dict(counter_result))

print("\n=== ì½”ë”©í…ŒìŠ¤íŠ¸ í•µì‹¬ íŒ¨í„´ ===")
print("ğŸ’¡ ë°°ì—´ ì¹´ìš´íŒ… íŒ¨í„´ (ê°€ì¥ ë¹ ë¦„):")
print("cnt = [0] * 26")
print("cnt[ord(c) - ord('a')] += 1")
print("â†’ ë¬¸ìë¥¼ 0~25 ì¸ë±ìŠ¤ë¡œ ë³€í™˜í•˜ì—¬ ì§ì ‘ ì¹´ìš´íŒ…")

print("\n=== ë¬¸ìì—´ ì „ì²˜ë¦¬ ===")
text = "  Hello\nWorld\t  "
print(f"ì›ë³¸: '{text}'")
print(f"strip(): '{text.strip()}'")           # ì–‘ìª½ ê³µë°± ì œê±°
print(f"rstrip('\\n'): '{text.rstrip()}'")    # ì˜¤ë¥¸ìª½ ê³µë°±/ê°œí–‰ ì œê±°
print(f"lower(): '{text.strip().lower()}'")   # ì†Œë¬¸ì ë³€í™˜

# ì‹¤ì œ ì½”ë”©í…ŒìŠ¤íŠ¸ì—ì„œ ìì£¼ ì‚¬ìš©ë˜ëŠ” ì…ë ¥ ì²˜ë¦¬
input_line = "Hello World\n"
processed = input_line.rstrip('\n')
print(f"ì…ë ¥ ì²˜ë¦¬: '{processed}'")
````

````python
# 1-5. ì„ í˜• ìŠ¤ìº” O(n) ì˜ˆì œ - êµ¬ê°„ í•©
arr = [1, 2, 3, 4, 5]

print("=== êµ¬ê°„ í•© (Prefix Sum) ===")
# ëˆ„ì  í•© ë°°ì—´ ë§Œë“¤ê¸°
prefix_sum = [0] * (len(arr) + 1)
for i in range(len(arr)):
    prefix_sum[i + 1] = prefix_sum[i] + arr[i] # prefix_sum = [0, 1, 3, 6, 10, 15]

print(f"ì›ë³¸ ë°°ì—´: {arr}")
print(f"ëˆ„ì  í•© ë°°ì—´: {prefix_sum}")

# êµ¬ê°„ [l, r]ì˜ í•©ì„ O(1)ì— êµ¬í•˜ê¸°
def range_sum(l, r):  # l, rì€ 0-indexed
    return prefix_sum[r] - prefix_sum[l - 1]

print(f"êµ¬ê°„ [1, 3]ì˜ í•©: {range_sum(1, 3)}")  # 2+3+4 = 9
print(f"êµ¬ê°„ [2, 5]ì˜ í•©: {range_sum(2, 5)}")  # 2+3+4+5 = 12

print("\n=== ì¹´ìš´íŒ…ìœ¼ë¡œ ìµœë¹ˆê°’ ì°¾ê¸° ===")
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
count = {}
max_count = 0
most_frequent = None

for num in data:
    count[num] = count.get(num, 0) + 1
    if count[num] > max_count:
        max_count = count[num]
        most_frequent = num

print(f"ë°ì´í„°: {data}")
print(f"ì¹´ìš´íŠ¸: {count}")
print(f"ìµœë¹ˆê°’: {most_frequent} (ë“±ì¥íšŸìˆ˜: {max_count})")
````

- ìŠ¤íƒì— **ì¸ë±ìŠ¤** vs **ê°’** ì¤‘ ë¬´ì—‡ì„ ë„£ì„ì§€ ë¯¸ë¦¬ ê²°ì •

````python
# 2-1. ê¸°ë³¸ ìŠ¤íƒ ì—°ì‚°
stack = []

print("=== ìŠ¤íƒ ê¸°ë³¸ ì—°ì‚° ===")
# push (append)
stack.append(1)
stack.append(2)
stack.append(3)
print(f"push í›„: {stack}")

# pop
top = stack.pop()
print(f"popëœ ê°’: {top}, ìŠ¤íƒ: {stack}")

# peek (top í™•ì¸)
if stack:  # ë¹ˆ ìŠ¤íƒ ì²´í¬ ì¤‘ìš”!
    print(f"top: {stack[-1]}")

# ìŠ¤íƒì´ ë¹„ì–´ìˆëŠ”ì§€ í™•ì¸
print(f"ìŠ¤íƒì´ ë¹„ì–´ìˆë‚˜? {len(stack) == 0}")

print("\n=== ì˜¬ë°”ë¥¸ ê´„í˜¸ ê²€ì‚¬ ===")
def is_valid_parentheses(s):
    stack = []
    pairs = {'(': ')', '[': ']', '{': '}'}
    
    for char in s:
        if char in pairs:  # ì—¬ëŠ” ê´„í˜¸
            stack.append(char)
        elif char in pairs.values():  # ë‹«ëŠ” ê´„í˜¸
            if not stack:  # ë¹ˆ ìŠ¤íƒ
                return False
            if pairs[stack.pop()] != char:  # ë§¤ì¹­ ì•ˆë¨
                return False
    
    return len(stack) == 0  # ëª¨ë“  ê´„í˜¸ê°€ ë§¤ì¹­ë˜ì–´ì•¼ í•¨

test_cases = ["()", "()[]{}", "([)]", "((", "))"]
for test in test_cases:
    result = is_valid_parentheses(test)
    print(f"'{test}': {result}")
````

````python
# 2-2. ë‹¨ì¡°ìŠ¤íƒ (Monotonic Stack) - ë‹¤ìŒ í° ì›ì†Œ ì°¾ê¸°
def next_greater_element(arr):
    """ê° ì›ì†Œì— ëŒ€í•´ ì˜¤ë¥¸ìª½ì—ì„œ ì²˜ìŒìœ¼ë¡œ ë‚˜íƒ€ë‚˜ëŠ” ë” í° ì›ì†Œë¥¼ ì°¾ìŒ"""
    n = len(arr)
    result = [-1] * n  # ê¸°ë³¸ê°’: -1 (ë” í° ì›ì†Œ ì—†ìŒ)
    stack = []  # ì¸ë±ìŠ¤ë¥¼ ì €ì¥
    
    for i in range(n):
        # í˜„ì¬ ì›ì†Œê°€ ìŠ¤íƒ topì˜ ì›ì†Œë³´ë‹¤ í¬ë©´ pop
        while stack and arr[stack[-1]] < arr[i]:
            idx = stack.pop()
            result[idx] = arr[i]
        stack.append(i)
    
    return result

arr = [2, 1, 2, 4, 3, 1]
nge = next_greater_element(arr)
print("=== ë‹¤ìŒ í° ì›ì†Œ ì°¾ê¸° ===")
print(f"ë°°ì—´: {arr}")
print(f"ê²°ê³¼: {nge}")

for i in range(len(arr)):
    if nge[i] == -1:
        print(f"{arr[i]} -> ì—†ìŒ")
    else:
        print(f"{arr[i]} -> {nge[i]}")

print("\n=== ë‹¨ì¡°ìŠ¤íƒ í…œí”Œë¦¿ ===")
def monotonic_stack_template(arr):
    stack = []
    for i, x in enumerate(arr):
        # ì¡°ê±´: ê°ì†Œí•˜ëŠ” ìŠ¤íƒ ìœ ì§€ (xë³´ë‹¤ í° ê°’ë“¤ì„ pop)
        while stack and stack[-1] > x:
            stack.pop()
        stack.append(x)
        print(f"i={i}, x={x}, stack={stack}")

print("ê°ì†Œí•˜ëŠ” ë‹¨ì¡°ìŠ¤íƒ:")
monotonic_stack_template([4, 3, 5, 1, 2])
````

# 3. íÂ·ë± (Queue & Deque)

## í•µì‹¬ ê°œë…
- **í (Queue)**: FIFO (First In, First Out) - ë¨¼ì € ë“¤ì–´ê°„ ê²ƒì´ ë¨¼ì € ë‚˜ì˜´
- **ë± (Deque)**: ì–‘ìª½ ëì—ì„œ ì‚½ì…/ì‚­ì œê°€ ëª¨ë‘ O(1)ì— ê°€ëŠ¥
- **ìš©ë„**: ì‹œë®¬ë ˆì´ì…˜, íšŒì „, BFSì˜ ê¸°ë°˜

## ì£¼ì˜ì‚¬í•­
- `list.pop(0)`ì€ O(n)ì´ë¯€ë¡œ ì‚¬ìš© ê¸ˆì§€! â†’ `deque.popleft()` ì‚¬ìš©
- íšŒì „ì€ `deque.rotate(k)` ì‚¬ìš©

````python
from collections import deque

print("=== ë±(Deque) ê¸°ë³¸ ì—°ì‚° ===")
dq = deque([1, 2])
print(f"ì´ˆê¸° ë±: {list(dq)}")

# ì˜¤ë¥¸ìª½ì— ì¶”ê°€/ì œê±°
dq.append(3)
print(f"append(3): {list(dq)}")

# ì™¼ìª½ì— ì¶”ê°€/ì œê±°
dq.appendleft(0)
print(f"appendleft(0): {list(dq)}")

# ì œê±° ì—°ì‚°
right = dq.pop()
left = dq.popleft()
print(f"pop(): {right}, popleft(): {left}")
print(f"í˜„ì¬ ë±: {list(dq)}")

print("\n=== íšŒì „ ì—°ì‚° ===")
dq = deque([1, 2, 3, 4, 5])
print(f"ì›ë³¸: {list(dq)}")

dq.rotate(2)  # ì˜¤ë¥¸ìª½ìœ¼ë¡œ 2ì¹¸ íšŒì „
print(f"rotate(2): {list(dq)}")

dq.rotate(-1)  # ì™¼ìª½ìœ¼ë¡œ 1ì¹¸ íšŒì „
print(f"rotate(-1): {list(dq)}")

print("\n=== í ì‹œë®¬ë ˆì´ì…˜ - í”„ë¦°í„° ===")
def printer_simulation(priorities, location):
    """íŠ¹ì • ë¬¸ì„œê°€ ì–¸ì œ ì¸ì‡„ë˜ëŠ”ì§€ ì‹œë®¬ë ˆì´ì…˜"""
    queue = deque()
    for i, priority in enumerate(priorities):
        queue.append((i, priority))  # (ì¸ë±ìŠ¤, ìš°ì„ ìˆœìœ„)
    
    count = 0
    while queue:
        current = queue.popleft()
        
        # í˜„ì¬ ë¬¸ì„œë³´ë‹¤ ìš°ì„ ìˆœìœ„ ë†’ì€ ë¬¸ì„œê°€ ìˆëŠ”ì§€ í™•ì¸
        if any(current[1] < doc[1] for doc in queue):
            queue.append(current)  # ë§¨ ë’¤ë¡œ ë³´ëƒ„
        else:
            count += 1
            if current[0] == location:
                return count
    
    return count

priorities = [2, 1, 3, 2]
location = 2
result = printer_simulation(priorities, location)
print(f"ìš°ì„ ìˆœìœ„: {priorities}, ì°¾ëŠ” ë¬¸ì„œ ìœ„ì¹˜: {location}")
print(f"ì¸ì‡„ ìˆœì„œ: {result}ë²ˆì§¸")
````

# 5. ì¢…í•© ì—°ìŠµ ë¬¸ì œ

ì‹¤ì œ ì½”ë”©í…ŒìŠ¤íŠ¸ì—ì„œ ë‚˜ì˜¬ ìˆ˜ ìˆëŠ” íŒ¨í„´ë“¤ì„ ì—°ìŠµí•´ë´…ì‹œë‹¤.

````python
# ì—°ìŠµ ë¬¸ì œ 1: ë‘ ìˆ˜ì˜ í•© (Two Sum) - í•´ì‹œ í™œìš©
def two_sum(nums, target):
    """ë°°ì—´ì—ì„œ ë‘ ìˆ˜ì˜ í•©ì´ targetì´ ë˜ëŠ” ì¸ë±ìŠ¤ ë°˜í™˜"""
    seen = {}  # ê°’ -> ì¸ë±ìŠ¤ ë§¤í•‘
    
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    
    return []

nums = [2, 7, 11, 15]
target = 9
result = two_sum(nums, target)
print(f"ë°°ì—´: {nums}, ëª©í‘œ: {target}")
print(f"ê²°ê³¼: ì¸ë±ìŠ¤ {result} -> {nums[result[0]]} + {nums[result[1]]} = {target}")

print("\n" + "="*50)

# ì—°ìŠµ ë¬¸ì œ 2: ìœ íš¨í•œ íŒ°ë¦°ë“œë¡¬ - ë¬¸ìì—´ + íˆ¬ í¬ì¸í„°
def is_palindrome(s):
    """ì•ŒíŒŒë²³ê³¼ ìˆ«ìë§Œ ê³ ë ¤í•˜ì—¬ íŒ°ë¦°ë“œë¡¬ ê²€ì‚¬"""
    # ì „ì²˜ë¦¬: ì•ŒíŒŒë²³ê³¼ ìˆ«ìë§Œ ë‚¨ê¸°ê³  ì†Œë¬¸ìë¡œ ë³€í™˜
    cleaned = ''.join(c.lower() for c in s if c.isalnum())
    
    # íˆ¬ í¬ì¸í„°ë¡œ ë¹„êµ
    left, right = 0, len(cleaned) - 1
    while left < right:
        if cleaned[left] != cleaned[right]:
            return False
        left += 1
        right -= 1
    
    return True

test_strings = ["A man, a plan, a canal: Panama", "race a car", ""]
for s in test_strings:
    result = is_palindrome(s)
    print(f"'{s}' -> {result}")

print("\n" + "="*50)
````

````python
# ì—°ìŠµ ë¬¸ì œ 3: ì¼ì¼ ì˜¨ë„ - ë‹¨ì¡°ìŠ¤íƒ í™œìš©
def daily_temperatures(temperatures):
    """ê° ë‚ ì— ëŒ€í•´ ë” ë”°ëœ»í•œ ë‚ ì´ ë‚˜ì˜¬ ë•Œê¹Œì§€ ê¸°ë‹¤ë ¤ì•¼ í•˜ëŠ” ì¼ìˆ˜"""
    n = len(temperatures)
    result = [0] * n
    stack = []  # ì¸ë±ìŠ¤ ì €ì¥
    
    for i in range(n):
        # í˜„ì¬ ì˜¨ë„ê°€ ìŠ¤íƒì˜ ë‚ ë“¤ë³´ë‹¤ ë†’ìœ¼ë©´
        while stack and temperatures[stack[-1]] < temperatures[i]:
            prev_day = stack.pop()
            result[prev_day] = i - prev_day  # ê¸°ë‹¤ë¦° ì¼ìˆ˜
        stack.append(i)
    
    return result

temps = [73, 74, 75, 71, 69, 72, 76, 73]
wait_days = daily_temperatures(temps)
print("ì¼ì¼ ì˜¨ë„:", temps)
print("ê¸°ë‹¤ë¦° ì¼ìˆ˜:", wait_days)
for i, (temp, wait) in enumerate(zip(temps, wait_days)):
    if wait == 0:
        print(f"  {i}ì¼ì°¨ ({temp}Â°): ë” ë”°ëœ»í•œ ë‚  ì—†ìŒ")
    else:
        print(f"  {i}ì¼ì°¨ ({temp}Â°): {wait}ì¼ í›„ ë” ë”°ëœ»í•´ì§")

print("\n" + "="*50)

# ì—°ìŠµ ë¬¸ì œ 4: ìŠ¬ë¼ì´ë”© ìœˆë„ìš° ìµœëŒ€ê°’ - ë± í™œìš©
from collections import deque

def sliding_window_maximum(nums, k):
    """í¬ê¸° kì¸ ìŠ¬ë¼ì´ë”© ìœˆë„ìš°ì˜ ìµœëŒ€ê°’ë“¤ì„ ë°˜í™˜"""
    if not nums or k == 0:
        return []
    
    dq = deque()  # ì¸ë±ìŠ¤ ì €ì¥ (ê°ì†Œí•˜ëŠ” ìˆœì„œ ìœ ì§€)
    result = []
    
    for i in range(len(nums)):
        # ìœˆë„ìš° ë²”ìœ„ë¥¼ ë²—ì–´ë‚œ ì¸ë±ìŠ¤ ì œê±°
        while dq and dq[0] <= i - k:
            dq.popleft()
        
        # í˜„ì¬ ê°’ë³´ë‹¤ ì‘ì€ ê°’ë“¤ì˜ ì¸ë±ìŠ¤ ì œê±°
        while dq and nums[dq[-1]] < nums[i]:
            dq.pop()
        
        dq.append(i)
        
        # ìœˆë„ìš°ê°€ ì™„ì„±ë˜ë©´ ìµœëŒ€ê°’ ì¶”ê°€
        if i >= k - 1:
            result.append(nums[dq[0]])
    
    return result

nums = [1, 3, -1, -3, 5, 3, 6, 7]
k = 3
max_values = sliding_window_maximum(nums, k)
print(f"ë°°ì—´: {nums}")
print(f"ìœˆë„ìš° í¬ê¸°: {k}")
print(f"ê° ìœˆë„ìš°ì˜ ìµœëŒ€ê°’: {max_values}")

# ìœˆë„ìš°ë³„ ìƒì„¸ ì¶œë ¥
for i in range(len(nums) - k + 1):
    window = nums[i:i+k]
    print(f"  ìœˆë„ìš° {window} -> ìµœëŒ€ê°’: {max_values[i]}")
````
